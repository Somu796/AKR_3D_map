```python
from typing import Tuple, Optional
import numpy as np
import xarray as xr

def create_3d_grid(
    x_range: Tuple[float, float] = (-15.0, 15.0),
    y_range: Tuple[float, float] = (-15.0, 15.0),
    z_range: Tuple[float, float] = (-10.0, 10.0),
    bin_size: float = 0.5
) -> xr.Dataset:
    """
    Create empty 3D grid around Earth.
    """
    
    # Type checking (optional but helpful during development)
    if not isinstance(x_range, tuple) or len(x_range) != 2:
        raise TypeError(f"x_range must be tuple of 2 floats, got {type(x_range)}")
    if not isinstance(bin_size, (int, float)) or bin_size <= 0:
        raise ValueError(f"bin_size must be positive number, got {bin_size}")
    
    # Create bin edges
    x_bins: np.ndarray = np.arange(x_range[0], x_range[1] + bin_size, bin_size)
    y_bins: np.ndarray = np.arange(y_range[0], y_range[1] + bin_size, bin_size)
    z_bins: np.ndarray = np.arange(z_range[0], z_range[1] + bin_size, bin_size)
    
    # Create bin centers
    x_centers: np.ndarray = (x_bins[:-1] + x_bins[1:]) / 2
    y_centers: np.ndarray = (y_bins[:-1] + y_bins[1:]) / 2
    z_centers: np.ndarray = (z_bins[:-1] + z_bins[1:]) / 2
    
    # Get dimensions
    n_x: int = len(x_centers)
    n_y: int = len(y_centers)
    n_z: int = len(z_centers)
    
    # Initialize arrays with proper dtype
    shape: Tuple[int, int, int] = (n_x, n_y, n_z)
    residence_time: np.ndarray = np.zeros(shape, dtype=np.float64)
    burst_count: np.ndarray = np.zeros(shape, dtype=np.int32)
    burst_time: np.ndarray = np.zeros(shape, dtype=np.float64)
    probability: np.ndarray = np.zeros(shape, dtype=np.float64)
    
    # Create Dataset
    grid: xr.Dataset = xr.Dataset(
        data_vars={
            'residence_time': (['x', 'y', 'z'], residence_time,
                              {'units': 'seconds', 'dtype': 'float64'}),
            'burst_count': (['x', 'y', 'z'], burst_count,
                           {'units': 'count', 'dtype': 'int32'}),
            'burst_time': (['x', 'y', 'z'], burst_time,
                          {'units': 'seconds', 'dtype': 'float64'}),
            'probability': (['x', 'y', 'z'], probability,
                          {'units': 'percent', 'dtype': 'float64'}),
        },
        coords={
            'x': (['x'], x_centers, {'units': 'R_E', 'dtype': 'float64'}),
            'y': (['y'], y_centers, {'units': 'R_E', 'dtype': 'float64'}),
            'z': (['z'], z_centers, {'units': 'R_E', 'dtype': 'float64'}),
        },
        attrs={
            'coordinate_system': 'GSE',
            'units': 'Earth_radii',
            'bin_size': float(bin_size),
            'description': 'AKR detection probability grid'
        }
    )
    
    return grid
```

```python
import plotly.graph_objects as go
import numpy as np

# Create the grid
grid = create_3d_grid(
    x_range=(-15.0, 15.0),
    y_range=(-15.0, 15.0),
    z_range=(-10.0, 10.0),
    bin_size=2.0
)

print(f"Grid created: {grid.dims}")
print(f"Shape: {grid.residence_time.shape}")

# Get bin edges from centers
bin_size = grid.attrs['bin_size']
x_edges = np.concatenate([grid.x.values - bin_size/2, [grid.x.values[-1] + bin_size/2]])
y_edges = np.concatenate([grid.y.values - bin_size/2, [grid.y.values[-1] + bin_size/2]])
z_edges = np.concatenate([grid.z.values - bin_size/2, [grid.z.values[-1] + bin_size/2]])

# Create figure
fig = go.Figure()

# Draw grid lines along X-axis
for y in y_edges:
    for z in z_edges:
        fig.add_trace(go.Scatter3d(
            x=x_edges, y=[y]*len(x_edges), z=[z]*len(x_edges),
            mode='lines', line=dict(color='gray', width=1),
            opacity=0.3, showlegend=False, hoverinfo='skip'
        ))

# Draw grid lines along Y-axis
for x in x_edges:
    for z in z_edges:
        fig.add_trace(go.Scatter3d(
            x=[x]*len(y_edges), y=y_edges, z=[z]*len(y_edges),
            mode='lines', line=dict(color='gray', width=1),
            opacity=0.3, showlegend=False, hoverinfo='skip'
        ))

# Draw grid lines along Z-axis
for x in x_edges:
    for y in y_edges:
        fig.add_trace(go.Scatter3d(
            x=[x]*len(z_edges), y=[y]*len(z_edges), z=z_edges,
            mode='lines', line=dict(color='gray', width=1),
            opacity=0.3, showlegend=False, hoverinfo='skip'
        ))

# Add Earth sphere at origin
u = np.linspace(0, 2*np.pi, 50)
v = np.linspace(0, np.pi, 50)
x_sphere = np.outer(np.cos(u), np.sin(v))
y_sphere = np.outer(np.sin(u), np.sin(v))
z_sphere = np.outer(np.ones(np.size(u)), np.cos(v))

fig.add_trace(go.Surface(
    x=x_sphere, y=y_sphere, z=z_sphere,
    colorscale='Blues', showscale=False,
    opacity=0.6, name='Earth'
))

# Add Sun (positioned along positive X-axis, typical GSE orientation)
sun_distance = -150  # Sun at ~150 R_E along -X axis (sunward direction)
sun_radius = 5  # Visual size

fig.add_trace(go.Surface(
    x=x_sphere * sun_radius + sun_distance,
    y=y_sphere * sun_radius,
    z=z_sphere * sun_radius,
    colorscale=[[0, 'yellow'], [1, 'orange']],
    showscale=False,
    opacity=0.8,
    name='Sun'
))

# Layout
fig.update_layout(
    title='3D Grid with Earth and Sun',
    scene=dict(
        xaxis_title='X (R_E)',
        yaxis_title='Y (R_E)',
        zaxis_title='Z (R_E)',
        aspectmode='data'
    ),
    width=900, height=800
)

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np
import xarray as xr
from typing import Optional


def visualize_grid_3d(
    grid: xr.Dataset,
    show_earth: bool = True,
    grid_opacity: float = 0.3,
    title: str = "3D Grid Visualization"
) -> go.Figure:
    """
    Visualize the 3D grid structure using Plotly.
    
    Parameters
    ----------
    grid : xr.Dataset
        Grid dataset from create_3d_grid()
    show_earth : bool
        Whether to show Earth sphere at origin
    grid_opacity : float
        Opacity of grid lines (0-1)
    title : str
        Plot title
    
    Returns
    -------
    go.Figure
        Plotly figure object
    """
    
    # Extract bin edges (we need edges, not centers, for grid lines)
    x_coords = grid.x.values
    y_coords = grid.y.values
    z_coords = grid.z.values
    
    # Get bin size
    bin_size = grid.attrs['bin_size']
    
    # Reconstruct edges from centers
    x_edges = np.concatenate([x_coords - bin_size/2, [x_coords[-1] + bin_size/2]])
    y_edges = np.concatenate([y_coords - bin_size/2, [y_coords[-1] + bin_size/2]])
    z_edges = np.concatenate([z_coords - bin_size/2, [z_coords[-1] + bin_size/2]])
    
    # Create figure
    fig = go.Figure()
    
    # === Draw grid lines ===
    
    # Lines parallel to X-axis
    for y in y_edges:
        for z in z_edges:
            fig.add_trace(go.Scatter3d(
                x=x_edges,
                y=[y] * len(x_edges),
                z=[z] * len(x_edges),
                mode='lines',
                line=dict(color='gray', width=1),
                opacity=grid_opacity,
                showlegend=False,
                hoverinfo='skip'
            ))
    
    # Lines parallel to Y-axis
    for x in x_edges:
        for z in z_edges:
            fig.add_trace(go.Scatter3d(
                x=[x] * len(y_edges),
                y=y_edges,
                z=[z] * len(y_edges),
                mode='lines',
                line=dict(color='gray', width=1),
                opacity=grid_opacity,
                showlegend=False,
                hoverinfo='skip'
            ))
    
    # Lines parallel to Z-axis
    for x in x_edges:
        for y in y_edges:
            fig.add_trace(go.Scatter3d(
                x=[x] * len(z_edges),
                y=[y] * len(z_edges),
                z=z_edges,
                mode='lines',
                line=dict(color='gray', width=1),
                opacity=grid_opacity,
                showlegend=False,
                hoverinfo='skip'
            ))
    
    # === Add Earth sphere (optional) ===
    if show_earth:
        # Create sphere mesh
        u = np.linspace(0, 2 * np.pi, 50)
        v = np.linspace(0, np.pi, 50)
        x_sphere = np.outer(np.cos(u), np.sin(v))
        y_sphere = np.outer(np.sin(u), np.sin(v))
        z_sphere = np.outer(np.ones(np.size(u)), np.cos(v))
        
        fig.add_trace(go.Surface(
            x=x_sphere,
            y=y_sphere,
            z=z_sphere,
            colorscale='Blues',
            showscale=False,
            opacity=0.6,
            name='Earth',
            hoverinfo='name'
        ))
    
    # === Layout settings ===
    fig.update_layout(
        title=dict(
            text=title,
            x=0.5,
            xanchor='center'
        ),
        scene=dict(
            xaxis=dict(title='X (R_E)', backgroundcolor="white", gridcolor="lightgray"),
            yaxis=dict(title='Y (R_E)', backgroundcolor="white", gridcolor="lightgray"),
            zaxis=dict(title='Z (R_E)', backgroundcolor="white", gridcolor="lightgray"),
            aspectmode='data',  # Equal aspect ratio
            camera=dict(
                eye=dict(x=1.5, y=1.5, z=1.2)
            )
        ),
        width=900,
        height=800,
        showlegend=True
    )
    
    return fig


# === Usage Example ===
if __name__ == "__main__":
    # Create grid
    grid = create_3d_grid(
        x_range=(-15.0, 15.0),
        y_range=(-15.0, 15.0),
        z_range=(-10.0, 10.0),
        bin_size=2.0  # Larger bin_size for clearer visualization
    )
    
    # Visualize
    fig = visualize_grid_3d(grid, show_earth=True, grid_opacity=0.4)
    fig.show()
    
    # Print grid info
    print(f"Grid dimensions: {grid.dims}")
    print(f"Grid shape: {grid.residence_time.shape}")
    print(f"Bin size: {grid.attrs['bin_size']} R_E")
```

```python
def create_3d_grid(
    x_range: Tuple[float, float] = (-15.0, 15.0),
    y_range: Tuple[float, float] = (-15.0, 15.0),
    z_range: Tuple[float, float] = (-10.0, 10.0),
    bin_size: float = 0.5
) -> xr.Dataset:
    """Create a 3D grid dataset with specified ranges and bin size."""
    x_bins = np.arange(x_range[0], x_range[1] + bin_size, bin_size)
    y_bins = np.arange(y_range[0], y_range[1] + bin_size, bin_size)
    z_bins = np.arange(z_range[0], z_range[1] + bin_size, bin_size)

    grid = xr.Dataset(
        {
            "intensity": (["x", "y", "z"], np.zeros((len(x_bins)-1, len(y_bins)-1, len(z_bins)-1)))
        },
        coords={
            "x": (x_bins[:-1] + x_bins[1:]) / 2,
            "y": (y_bins[:-1] + y_bins[1:]) / 2,
            "z": (z_bins[:-1] + z_bins[1:]) / 2,
        }
    )
    return grid
```
